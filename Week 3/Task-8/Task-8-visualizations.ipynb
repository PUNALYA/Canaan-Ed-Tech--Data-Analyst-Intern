```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\punal\OneDrive\Desktop\Canaan Ed & Tech Internship\Week 3\Task-7\Task-7-sales_data.csv")
df.head()
```

```python
# Bar Chart
plt.bar(df["Product"], df["Price"])
plt.title("Product vs Price")
plt.xticks(rotation=45)
plt.savefig(r"C:\Users\punal\OneDrive\Desktop\Canaan Ed & Tech Internship\Week 3\Task-8\Charts\bar_chart.png")
plt.show()
```

```python
# Histogram
plt.hist(df["Price"], bins=5)
plt.title("Price Distribution")
plt.savefig(r"C:\Users\punal\OneDrive\Desktop\Canaan Ed & Tech Internship\Week 3\Task-8\Charts\histogram.png")
plt.show()
```

```python
# Line PLot
plt.plot(df["Product"], df["Quantity"])
plt.title("Product vs Quantity")
plt.xticks(rotation=45)
plt.savefig(r"C:\Users\punal\OneDrive\Desktop\Canaan Ed & Tech Internship\Week 3\Task-8\Charts\line_plot.png")
plt.show()
```

```python
# Scatter Plot
plt.scatter(df["Price"], df["Quantity"])
plt.title("Price vs Quantity")
plt.xlabel("Price")
plt.ylabel("Quantity")
plt.savefig(r"C:\Users\punal\OneDrive\Desktop\Canaan Ed & Tech Internship\Week 3\Task-8\Charts\scatter_plot.png")
plt.show()
```

```python
# Heatmap
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.savefig(r"C:\Users\punal\OneDrive\Desktop\Canaan Ed & Tech Internship\Week 3\Task-8\Charts\heatmap.png")
plt.show()
```
